```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("data.csv")
data.head(5)
```

```python
data.drop(columns=["id"]).head()
```

```python
sns.countplot(x="diagnosis",data=data)
```

```python
corr_matrix = data.corr()
plt.figure(figsize=(30,22))
plt.title("Correlation Between Features")
sns.heatmap(corr_matrix, annot=True, fmt=".3f")
```

```python
data['diagnosis'] = data['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
data.head()
```

```python
corr_matrix=["diagnosis",'radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean']
data[corr_matrix].corr()["diagnosis"].sort_values()
```

```python
top_corr = ["diagnosis",'concave points_mean','perimeter_mean','radius_mean','area_mean','concavity_mean','compactness_mean','texture_mean']
sns.pairplot(data[top_corr],hue="diagnosis")
plt.show()
```

```python
data.hist(top_corr,figsize=(10,10))
```

```python
p=['area_mean','compactness_mean','texture_mean']
sns.boxplot(x='diagnosis',y='concave points_worst',data=data)
plt.show()
```
